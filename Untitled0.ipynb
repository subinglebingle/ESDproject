```python
import os
import random
import shutil

# 경로 설정
img_dir = '/content/dataset/data_resize'
label_dir = '/content/dataset/data_label'

out_img_train = '/content/dataset/images/train'
out_img_val = '/content/dataset/images/val'
out_label_train = '/content/dataset/labels/train'
out_label_val = '/content/dataset/labels/val'

# 출력 디렉토리 생성
for d in [out_img_train, out_img_val, out_label_train, out_label_val]:
    os.makedirs(d, exist_ok=True)

# 파일 리스트 로드
img_files = [f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]
random.shuffle(img_files)

# 80% 학습 / 20% 검증
split_idx = int(len(img_files) * 0.8)
train_imgs = img_files[:split_idx]
val_imgs = img_files[split_idx:]

def move_files(img_list, img_dst, label_dst):
    for img_name in img_list:
        base_name = os.path.splitext(img_name)[0]
        label_name = base_name + '.txt'

        # 이미지 복사
        shutil.copy(os.path.join(img_dir, img_name), os.path.join(img_dst, img_name))

        # 라벨이 있을 때만 복사
        label_path = os.path.join(label_dir, label_name)
        if os.path.exists(label_path):
            shutil.copy(label_path, os.path.join(label_dst, label_name))
        else:
            print(f"⚠️ 라벨 없음: {label_path}")

# 파일 이동
move_files(train_imgs, out_img_train, out_label_train)
move_files(val_imgs, out_img_val, out_label_val)

print("✅ 데이터 분할 완료!")
```
